```python
"""
Shift sine wave code to be shared with era splitting paper
"""

# Import the necessary modules and libraries
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2)

era_samples = 64
num_eras = 1

Xs = []
Ys1 = []
Ys2 = []
Ys3 = []

for i in range(num_eras):
    Xs.append(2 * np.pi * np.random.rand(era_samples, 1))
    Ys1.append(np.sin(Xs[-1]).ravel())
    Ys2.append(np.sin(Xs[-1]).ravel() + np.random.rand() * 2)
    Ys3.append(np.sin(Xs[-1]).ravel() + np.random.rand() * 2 + 1 * np.random.randn(era_samples))

X_train = np.vstack(Xs)
y_train1 = np.hstack(Ys1)
y_train2 = np.hstack(Ys2)
y_train3 = np.hstack(Ys3)

X_test = np.sort(2 * np.pi * np.random.rand(era_samples, 1), axis=0)
y_test = np.sin(X_test).ravel()

eras = np.concatenate([(np.ones(era_samples)).astype('int') * n for n in range(num_eras)]) + 1

fig, axs = plt.subplots(1, 3, figsize=(12, 4))  # Create subplots

# Plotting y_train1
axs[0].scatter(X_train, y_train1, s=20, edgecolor="black", c="darkorange", label="data", alpha=0.5)
axs[0].scatter(X_test, y_test, s=20, edgecolor="black", c="darkgreen", label="target", alpha=0.3)
axs[0].set_xlabel("input data")
axs[0].set_ylabel("target")
axs[0].set_title("Step 1")
axs[0].legend()

# Plotting y_train2
axs[1].scatter(X_train, y_train2, s=20, edgecolor="black", c="darkorange", label="data", alpha=0.5)
axs[1].scatter(X_test, y_test, s=20, edgecolor="black", c="darkgreen", label="target", alpha=0.3)
axs[1].set_xlabel("input data")
axs[1].set_ylabel("target")
axs[1].set_title("Step 2")
axs[1].legend()

# Plotting y_train3
axs[2].scatter(X_train, y_train3, s=20, edgecolor="black", c="darkorange", label="data", alpha=0.5)
axs[2].scatter(X_test, y_test, s=20, edgecolor="black", c="darkgreen", label="target", alpha=0.3)
axs[2].set_xlabel("input data")
axs[2].set_ylabel("target")
axs[2].set_title("Step 3")
axs[2].legend()

plt.tight_layout()  # Adjust subplot spacing
plt.savefig('step123.png')
plt.show()
```

```python
"""
Experiment Data
"""

# Import the necessary modules and libraries
import numpy as np
import matplotlib.pyplot as plt

# np.random.seed(4)


num_eras = 64
era_samples = 32

shift_magnitude_multiplier = 2
blur_magnitude = 2


Xs = []
Ys = []
eras = []

for i in range(num_eras):
    Xs.append(2 * np.pi * np.random.rand(era_samples, 1))
    Ys.append(np.sin(Xs[-1]).ravel() + np.random.randn() * shift_magnitude_multiplier + blur_magnitude * np.random.randn(era_samples))
    eras.append([i]*era_samples)

X_train = np.vstack(Xs)
y_train = np.hstack(Ys)
eras_train = np.hstack(eras)

Xs = []
Ys = []

for i in range(num_eras):
    Xs.append(2 * np.pi * np.random.rand(era_samples, 1))
    Ys.append(np.sin(Xs[-1]).ravel() + np.random.randn() * shift_magnitude_multiplier + blur_magnitude * np.random.randn(era_samples))

X_test = np.vstack(Xs)
y_test = np.hstack(Ys)
signal = np.sin( X_test )

fig, axs = plt.subplots(1, 1, figsize=(8, 6))  # Create subplots

# Plotting y_train1
axs.scatter(X_train, y_train, s=20, edgecolor="black", c="darkorange", label="Training Data", alpha=0.5)
axs.scatter(X_test, y_test, s=20, edgecolor="black", c="darkgreen", label="Test Data", alpha=0.3)
axs.set_xlabel("input data")
axs.set_ylabel("target")
axs.set_title(f"Test Data: Num Eras {num_eras} ")
axs.legend()

plt.tight_layout()  # Adjust subplot spacing
plt.savefig('Experimental_Data.png')
plt.show()

fig, axs = plt.subplots(1, 1, figsize=(8, 6))  # Create subplots

# Plotting y_train1
axs.scatter(X_test, signal, s=20, edgecolor="black", c="darkgreen", label="Signal Data", alpha=0.3)
axs.set_xlabel("input data")
axs.set_ylabel("target")
axs.set_title(f"Signal Data ")
axs.legend()

plt.tight_layout()  # Adjust subplot spacing
plt.savefig('Signal_Data.png')
plt.show()
```

```python
'''
Era Splitting
'''
import numpy as np
import pandas as pd
from sklearn.ensemble import EraHistGradientBoostingRegressor
import lightgbm as lgbm

baseline_model = lgbm.LGBMRegressor(
    n_estimators=100,
    max_depth=10,
    learning_rate=1
)
experimental_model = EraHistGradientBoostingRegressor(
    max_iter=100,
    max_depth=10,
    learning_rate=1,
    early_stopping=False,
    boltzmann_alpha=0,
    blama=0
)

baseline_model.fit( X_train, y_train)
experimental_model.fit( X_train, y_train, eras=eras_train)

baseline_preds = baseline_model.predict( X_test )
experimental_preds = experimental_model.predict( X_test )


sort_order = np.argsort(X_test.flatten())

baseline_mse = np.round( np.mean( (baseline_preds - y_test)**2 ), 2)
experimental_mse = np.round( np.mean( (experimental_preds - y_test)**2 ), 2)

baseline_mse_signal = np.round( np.mean( (baseline_preds - signal)**2 ), 2)
experimental_mse_signal = np.round( np.mean( (experimental_preds - signal)**2 ), 2)

baseline_corr = np.round( np.corrcoef([baseline_preds, y_test])[0,1], 2)
experimental_corr = np.round( np.corrcoef([experimental_preds, y_test])[0,1], 2)

baseline_corr_signal = np.round( np.corrcoef([baseline_preds, signal.flatten()])[0,1], 2)
experimental_corr_signal = np.round( np.corrcoef([experimental_preds, signal.flatten()])[0,1], 2)

plt.plot( X_test[sort_order], baseline_preds[sort_order], label=f"Baseline" )
plt.plot( X_test[sort_order], experimental_preds[sort_order], label=f"Experimental", linewidth=2 )


plt.xlabel('Input Data')
plt.ylabel('Target/Prediction')
plt.legend()
plt.title("Sine Wave w/ Era Splitting")
plt.savefig("sine_wave_era_splitting.png")

plt.show()


pd.DataFrame({
    "Target":["Test Data", "Test Data", "Signal", "Signal"],
    "Experiment":["Baseline","Experimental","Baseline","Experimental",],
    "MSE":[baseline_mse,experimental_mse, baseline_mse_signal, experimental_mse_signal],
    "Corr":[baseline_corr,experimental_corr, baseline_corr_signal, experimental_corr_signal],
}).set_index(["Target","Experiment"])
```
