```python
import pandas as pd
# from lightgbm import LGBMRegressor
import gc
import json
from pathlib import Path
from numerapi import NumerAPI

"""Era Split Model"""
from sklearn.ensemble import EraHistGradientBoostingRegressor

'''Baseline Model'''
from sklearn.ensemble import HistGradientBoostingRegressor

# download all the things

napi = NumerAPI()
# current_round = napi.get_current_round()

# Tournament data changes every week so we specify the round in their name. Training
# and validation data only change periodically, so no need to download them every time.
print('Downloading dataset files...')

Path("./v4").mkdir(parents=False, exist_ok=True)
# napi.download_dataset("v4/train.parquet")
# napi.download_dataset("v4/validation.parquet")
# napi.download_dataset("v4/live.parquet", f"v4/live_{current_round}.parquet")
# napi.download_dataset("v4/validation_example_preds.parquet")
# napi.download_dataset("v4/features.json")

print('Reading minimal training data')
# read the feature metadata and get a feature set (or all the features)
# with open("v4/features.json", "r") as f:
#     feature_metadata = json.load(f)
# features = feature_metadata["feature_sets"]['small']

```

```python
TARGET_COL="target_cyrus_v4_20"
# read_columns = features + ['era', TARGET_COL]

# note: sometimes when trying to read the downloaded data you get an error about invalid magic parquet bytes...
# if so, delete the file and rerun the napi.download_dataset to fix the corrupted file
training_data = pd.read_parquet('v4/train.parquet')
validation_data = pd.read_parquet('v4/validation.parquet')
# live_data = pd.read_parquet(f'v4/live_{current_round}.parquet')

features = [ f for f in list(training_data) if 'feature' in f ]

# pare down the number of eras to every 4th era
# every_4th_era = training_data['era'].unique()[::4]
# training_data = training_data[ training_data['era'].isin(every_4th_era) ]

# every_4th_era = validation_data['era'].unique()[::4]
# validation_data = validation_data[ validation_data['era'].isin(every_4th_era) ]

"""Baseline Model"""

model = HistGradientBoostingRegressor( 
    early_stopping=False, 
#     boltzmann_alpha=0, 
    max_iter=2000, 
    max_depth=5, 
    learning_rate=.01, 
    colsample_bytree=.1, 
    max_leaf_nodes=32, 
#     gamma=0.5, 
#     blama=0.5, 
#     era_boosting=False,
#     gain_debug=False,
#     vanna=0
)

# train on all of train and save the model so we don't have to train next time
model.fit(training_data[ features ], training_data[ TARGET_COL ])

validation_data.loc[:, f"preds_baseline"] = model.predict( validation_data[features] )

baseline_corrs = validation_data.groupby('era').apply( 
    lambda x: x[["preds_baseline", TARGET_COL]].corr().iloc[0,1] 
)

baseline_corrs.cumsum().plot()
desc = baseline_corrs.describe()
desc['sharpe'] = baseline_corrs.mean()/baseline_corrs.std()
desc['win_rate'] = ( baseline_corrs.dropna() > 0 ).sum() / len(baseline_corrs)
baseline_desc = desc


```

```python
'''Era Split'''

training_data['era'] = training_data['era'].astype('int')

era_model = EraHistGradientBoostingRegressor( 
    early_stopping=False, 
    boltzmann_alpha=0, 
    max_iter=2000, 
    max_depth=5, 
    learning_rate=.01, 
    colsample_bytree=.1, 
    max_leaf_nodes=32, 
    gamma=0, 
    blama=0, 
    era_boosting=False,
    gain_debug=False,
    vanna=0
)

era_model.fit(training_data[ features ], training_data[ TARGET_COL ], training_data['era'].values)
validation_data.loc[:, f"preds_era_split"] = era_model.predict( validation_data[features] )

era_split_corrs = validation_data.groupby('era').apply( 
    lambda x: x[["preds_era_split", TARGET_COL]].corr().iloc[0,1] 
)

era_split_corrs.cumsum().plot()
desc = era_split_corrs.describe()
desc['sharpe'] = era_split_corrs.mean()/era_split_corrs.std()
desc['win_rate'] = ( era_split_corrs.dropna() > 0 ).sum() / len(era_split_corrs.dropna())
pd.DataFrame(  desc )
erasplit_desc = desc
```

```python
'''Era Split Mixin'''

training_data['era'] = training_data['era'].astype('int')
training_data['new_era'] = ( training_data['era'] ).round().astype('int')

era_model = EraHistGradientBoostingRegressor( 
    early_stopping=False, 
    boltzmann_alpha=0, 
    max_iter=2000, 
    max_depth=5, 
    learning_rate=.01, 
    colsample_bytree=.1, 
    max_leaf_nodes=32, 
    gamma=0.5, 
    blama=0, 
    era_boosting=False,
    gain_debug=False,
    vanna=0
)

era_model.fit(training_data[ features ], training_data[ TARGET_COL ], training_data['era'].values)
validation_data.loc[:, f"preds_era_split"] = era_model.predict( validation_data[features] )



era_split_corrs = validation_data.groupby('era').apply( 
    lambda x: x[["preds_era_split", TARGET_COL]].corr().iloc[0,1] 
)

era_split_corrs.cumsum().plot()
desc = era_split_corrs.describe()
desc['sharpe'] = era_split_corrs.mean()/era_split_corrs.std()
desc['win_rate'] = ( era_split_corrs.dropna() > 0 ).sum() / len(era_split_corrs)
pd.DataFrame(  desc )
erasplit_mixin_desc = desc
```

```python
'''Directional Era Split'''

training_data['era'] = training_data['era'].astype('int')

era_model = EraHistGradientBoostingRegressor( 
    early_stopping=False, 
    boltzmann_alpha=0, 
    max_iter=2000, 
    max_depth=5, 
    learning_rate=.01, 
    colsample_bytree=.1, 
    max_leaf_nodes=32, 
    gamma=0, 
    blama=1, 
    era_boosting=False,
    gain_debug=False,
    vanna=0
)

era_model.fit(training_data[ features ], training_data[ TARGET_COL ], training_data['era'].values)
validation_data.loc[:, f"preds_era_split"] = era_model.predict( validation_data[features] )



era_split_corrs = validation_data.groupby('era').apply( 
    lambda x: x[["preds_era_split", TARGET_COL]].corr().iloc[0,1] 
)

era_split_corrs.cumsum().plot()
desc = era_split_corrs.describe()
desc['sharpe'] = era_split_corrs.mean()/era_split_corrs.std()
desc['win_rate'] = ( era_split_corrs.dropna() > 0 ).sum() / len(era_split_corrs.dropna())
pd.DataFrame(  desc )
dir_erasplit_desc = desc
```

```python
'''Directional Era Split Mixin'''

training_data['era'] = training_data['era'].astype('int')


era_model = EraHistGradientBoostingRegressor( 
    early_stopping=False, 
    boltzmann_alpha=0, 
    max_iter=2000, 
    max_depth=5, 
    learning_rate=.01, 
    colsample_bytree=.1, 
    max_leaf_nodes=32, 
    gamma=0.5, 
    blama=0.5, 
    era_boosting=False,
    gain_debug=False,
    vanna=0
)

era_model.fit(training_data[ features ], training_data[ TARGET_COL ], training_data['era'].values)
validation_data.loc[:, f"preds_era_split"] = era_model.predict( validation_data[features] )



era_split_corrs = validation_data.groupby('era').apply( 
    lambda x: x[["preds_era_split", TARGET_COL]].corr().iloc[0,1] 
)

era_split_corrs.cumsum().plot()
desc = era_split_corrs.describe()
desc['sharpe'] = era_split_corrs.mean()/era_split_corrs.std()
desc['win_rate'] = ( era_split_corrs.dropna() > 0 ).sum() / len(era_split_corrs.dropna())
pd.DataFrame(  desc )
dir_erasplit_mixin_desc = desc
```

```python
desc = pd.DataFrame( baseline_desc, columns=['Baseline'] )
desc['Era Split'] = erasplit_desc
desc['Era Split Mixin'] = erasplit_mixin_desc
desc['Directional Era Split'] = dir_erasplit_desc
desc['Directional Era Split Mixin'] = erasplit_mixin_desc
```
